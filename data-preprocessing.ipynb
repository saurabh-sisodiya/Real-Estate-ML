```python
df = df[df['price'] != 'Price on Request']
```

```python
def treat_price(x):
    if type(x) == float:
        return x

    parts = x.split(' ')
    if len(parts) < 2:
        return None
        
    if parts[1] == 'Lac':
        return round(float(x[0])/100,2)
    elif parts[1] == 'Crore':
        return round(float(x[0]),2)
    else:
        return None
```

```python
df['price'] = df['price'].apply(treat_price)
```
